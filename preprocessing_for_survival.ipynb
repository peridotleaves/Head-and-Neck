```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("normalized-rna-data.csv", sep=';')
```

```python
for c in df.columns[1:]:
    df[c] = df[c].str.replace(',','.')
    df[c] = pd.to_numeric(df[c])
```

```python
for r in range(len(df)):
    df.iloc[r,1:].fillna((df.iloc[r,1:].mean()), inplace=True)
```

```python
df.Gene = df.Gene.str.split('|')
genes = pd.DataFrame(df["Gene"].to_list(), columns=['Name', 'ID'])
```

```python
genes
```

```python
df.drop(['Gene'], axis=1, inplace=True)
```

```python
df_clean = pd.concat([genes, df], axis=1)
df_clean
```

```python
df_clean.to_csv("expression_clean.csv", index=False)
```
